```python
import csv
import numpy
```

```python
dd=[]
tm=[]
tx=[]
tn=[]
file='produkt_klima_tag_19510101_20231231_04336.txt'
f = open(file, 'r')
header = f.readline()
for line in f:
    line = line.strip()
    columns = line.split(';')
    dd.append(int(columns[1]))
    tm.append(float(columns[13]))
    tx.append(float(columns[15]))
    tn.append(float(columns[16]))
f.close()

file='produkt_klima_tag_20230701_20241231_04336.txt'
f = open(file, 'r')
header = f.readline()
for line in f:
    line = line.strip()
    columns = line.split(';')
    ddt=int(columns[1])
    if ddt>20231231:
        dd.append(int(columns[1]))
        tm.append(float(columns[13]))
        tx.append(float(columns[15]))
        tn.append(float(columns[16]))
f.close()
```

```python
for it in range(len(dd)):
    if tx[it]<0:
        print(dd[it],tx[it])
```

```python
header=['date','Tmean','Tmax','Tmin']
f = open('temp_tag_19510101_20241231_04336.csv', 'w', newline='')
writer = csv.writer(f)
writer.writerow(header)
for it in range(len(dd)):
    row=[dd[it],tm[it],tx[it],tn[it]]
    writer.writerow(row)
f.close()
```
